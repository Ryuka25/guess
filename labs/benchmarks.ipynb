```python
def show_heatmap(results):
    plt.imshow(results, origin="lower")

    cbar = plt.colorbar(label="Result Value")
    # Get the existing ticks
    current_ticks = cbar.get_ticks().tolist()

    # Add min and max if they are not already in the list
    new_ticks = sorted(set(current_ticks + [results.min(), results.max()]))

    cbar.set_ticks(new_ticks)
    plt.xlabel("b")
    plt.ylabel("a")
    plt.title("Heatmap of Result(a, b)")
    plt.show()
```

```python
def show_plots(x, y):
    # Create the plot
    plt.plot(x, y, marker='o', linestyle='-')

    # Add labels and title
    plt.xlabel("X value")
    plt.ylabel("Possibilities")
    plt.title("Possibilities per K values")

    # Display the plot
    plt.show()
```

```python
possibles_solutions = []
k_values = []

for i in range(1, 100 + 1):
    # numbers = {"K": 10,"L": 20, "X": 32}
    numbers = {"K": i,"L": 50}
    problem = Problem()
    problem.addVariables(numbers.keys(), range(1, 100 + 1))
    add_constraint_for_zero(problem, numbers, "K", "L")
    possibles_solutions.append(len(problem.getSolutions()))
    k_values.append(i)

show_plots(k_values, possibles_solutions)
```

```python
a = np.arange(0, 100 + 1)
b = np.arange(0, 100 + 1)

possibles_solutions = np.zeros((len(a), len(b)))

for i in range(len(a)):
    for j in range(len(b)):
        numbers = {"K": i,"L": j}
        problem = Problem()
        problem.addVariables(numbers.keys(), range(1, 100 + 1))
        add_constraint_for_zero(problem, numbers, "K", "L")
        possibles_solutions[i,j] = len(problem.getSolutions())

show_heatmap(possibles_solutions)
```
